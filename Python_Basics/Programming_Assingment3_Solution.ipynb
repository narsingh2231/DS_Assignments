```python
# 1.	Write a Python Program to Check if a Number is Positive, Negative or Zero?
num = float(input('please Enter a number :' ))
if (num > 0):
  print("It's a Positive Number")
elif (num < 0):
  print("It's a Negative Number")
else:
  print("It's a Zero")
```

```python
# 2.	Write a Python Program to Check if a Number is Odd or Even?
num = float(input('please Enter a number :' ))
mod = num % 2    #remainder shuold be zero
if (mod == 0):
  print("It's an Even Number")
else:
  print("\n It's an odd Number")
```

```python
# 3.	Write a Python Program to Check Leap Year?
Year = float(input('please Enter a Year(eg:2021) :' ))
mod = Year % 4    #It should be divisible by 4
if (mod == 0):
  print("\n It's a Leap Year")
else:
  print("\n It's not a Leap Year")
```

```python
# 4.	Write a Python Program to Check Prime Number?

num = int(input('please Enter a number :' ))
if ((num ==2) or (num==3) or (num== 5)):
  print("\n It's a Prime Number")
elif ((((num%2) != 0) and ((num%3)!= 0) and ((num%5) != 0))):           # It should be divisible by 2 or 3 or 5
  print("\n It's a Prime Number")
else:
  print("\n It's not Prime Number")
```

```python
# 5.	Write a Python Program to Print all Prime Numbers in an Interval of 1-10000?
for num in range(1,10000):
  if ((num ==2) or (num==3) or (num== 5)):
    print(num , end =" ")
  elif ((((num%2) != 0) and ((num%3)!= 0) and ((num%5) != 0))):           # It should be divisible by 2 or 3 or 5
    print(num, end =" ")
```
